3.Use z-score normalization to standardize the values of the income attribute. [Do not change the original income attribute in the table.]

```python
# Create Z score normalization 
income_zscore = (df["income"] - df["income"].mean())/ df["income"].std()
# Show 10 results
income_zscore.head(10)
```

4.Discretize the age attribute into 3 categories (corresponding to "young", "mid-age", and "old"). [Do not change the original age attribute in the table.]

```python
# Discretize the age into 3 categories
df_bins = pd.qcut(df.income, 3)
df_bins.head(10)
```

```python
# Bin the income variable and print first few
df_bins = pd.qcut(df.income, [0, .33, .66, 1], labels=["young", "mid-age", "old"])
df_bins.head(10)
```

5.Use Min-Max Normalization to transform the values of all numeric attributes  (income, age, children) in the original table (before the transformations in parts 3 and 4 above) onto the range 0.0-1.0.

```python
# Compute min/max of income - age - chi
min_inc = df["income"].min()
max_inc = df["income"].max()
min_age = df["age"].min()
max_age = df["age"].max()
min_chi = df["children"].min()
max_chi = df["children"].max()
print (min_inc, max_inc, min_age, max_age, min_chi, max_chi)
```

```python
#min-max normalization
norm_inc = (df["income"] - min_inc) / (max_inc - min_inc)
norm_age = (df["age"] - min_age) / (max_age - min_age)
norm_chi = (df["children"] - min_chi) / (max_chi - min_chi)
print (norm_inc.head(5))
print (norm_age.head(5))
print (norm_chi.head(5))
```

```python
# Transform to original table 

df["norm_inc"] = norm_inc
df["norm_age"] = norm_age
df["norm_chi"] = norm_chi
df["df_bins"] = df_bins
df["income_zscore"] = income_zscore
df.head(15)
```
